```python
# Check for missing values
df.isnull().sum()
```

```python
# Convert necessary columns to categorical data types
categorical_columns = ['ANSWERED', 'FEMALE', 'JOB', 'RENT', 'OWN_RES', 'NEW_CAR', 'CHK_ACCT', 'SAV_ACCT', 'MOBILE', 'PRODUCT']
df[categorical_columns] = df[categorical_columns].astype('category')
```

```python
# Plot income distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['INCOME'], kde=True)
plt.title("Income Distribution")
plt.xlabel("Income (USD)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot age distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['AGE'], kde=True)
plt.title("Age Distribution")
plt.xlabel("Age (years)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot income vs. answered
plt.figure(figsize=(8, 5))
sns.boxplot(x='ANSWERED', y='INCOME', data=df)
plt.title("Income vs. Call Response")
plt.xlabel("Call Answered")
plt.ylabel("Income (USD)")
plt.show()
```

```python
# Plot job type vs. answered
plt.figure(figsize=(8, 5))
sns.countplot(x='JOB', hue='ANSWERED', data=df)
plt.title("Job Type vs. Call Response")
plt.xlabel("Job Type")
plt.ylabel("Count")
plt.legend(title="Answered")
plt.show()
```

```python
from scipy.stats import ttest_ind

# Split income based on call response
income_answered = df[df['ANSWERED'] == 1]['INCOME']
income_not_answered = df[df['ANSWERED'] == 0]['INCOME']

# Perform t-test
t_stat, p_value = ttest_ind(income_answered, income_not_answered, equal_var=False)
print(f"T-test for INCOME effect on ANSWERED: t-stat={t_stat}, p-value={p_value}")
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table_female = pd.crosstab(df['ANSWERED'], df['FEMALE'])

# Perform chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table_female)
print(f"Chi-square test for FEMALE effect on ANSWERED: chi2={chi2}, p-value={p}")
```
